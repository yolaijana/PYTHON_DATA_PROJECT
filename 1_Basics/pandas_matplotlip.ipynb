```python
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
```

```python
df = dataset['train'].to_pandas()
```

```python
df
```

```python
df.drop(labels='salary_hour_avg', axis=1, inplace=True)
```

```python
df.info()
```

```python
import pandas as pd
```

```python
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df.info()
```

```python
df.sort_values(by='job_posted_date', inplace=True)
```

```python
us_jobs = df[df['job_country'] == 'United States']
```

```python
us_jobs = us_jobs[us_jobs['salary_year_avg'].notna()]
```

```python
us_jobs.groupby('job_title_short')['salary_year_avg'].count()
```

```python
 us_jobs.groupby('job_title_short').count()
```

```python
us_jobs.groupby('job_title_short')['salary_year_avg'].min()
```

```python
us_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median','min','max','count']).sort_values(by='median', ascending=False)
```

```python
job_count = df.job_title_short.value_counts()
```

```python
job_count
```

```python
plt.bar(job_count.index, job_count)
plt.title('Postings by Job Title')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
job_count = job_count.sort_values(ascending=[False])
plt.barh(job_count.index, job_count)
plt.show()
```
